```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("quote_dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
quotes = df['quote']
quotes.head()
```

```python
quotes = quotes.str.lower()
```

```python
import string
translator = str.maketrans('', '', string.punctuation)
quotes = quotes.apply(lambda x : x.translate(translator))
```

```python
quotes.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size = 10000
tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(quotes)
```

```python
word_index = tokenizer.word_index
print(len(word_index))
list(word_index.items())[:10]
```

```python
sequence = tokenizer.texts_to_sequences(quotes)
```

```python
quotes[0]
```

```python
sequence[0]
```

```python
X = []
y= []
for seq in sequence:
  for i in range(1,len(seq)):
    input_seq = seq[:i]
    output_seq = seq[i]
    X.append(input_seq)
    y.append(output_seq)
```

```python
X
```

```python
y
```

```python
len(X)
```

```python
len(y)
```

```python
max_len = max(len(x) for x in X)
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_padded = pad_sequences(X, maxlen=max_len, padding='pre')
```

```python
X_padded
```

```python
y = np.array(y)
```

```python
X_padded.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y_one_hot = to_categorical(y,num_classes = vocab_size)
```

```python
y.shape
```

```python
y_one_hot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense
```

```python
embedding_dim = 50
rnn_units = 128
```

```python
rnn_model = Sequential()
rnn_model.add(
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len)

)

rnn_model.add(SimpleRNN(units=rnn_units))
rnn_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
rnn_model.summary()
```

```python
lstm_model = Sequential()
lstm_model.add(
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len)
)
lstm_model.add(LSTM(units=rnn_units))
lstm_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
lstm_model.summary()
```

```python
epochs = 10
batch_size = 128
```

```python
history_rnn = rnn_model.fit(
    X_padded,
    y_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.1
)
```

```python
epochs = 100
batch_size = 128

history_lstm = lstm_model.fit(
    X_padded,
    y_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.1
)
```

```python
lstm_model.save("lstm_model.h5")
```

```python
index_to_word = {}
for word, index in word_index.items():
  index_to_word[index] = word
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def predictor(model, tokenizer,text, max_len):
  text= text.lower()
  seq = tokenizer.texts_to_sequences([text])[0]
  seq = pad_sequences([seq], maxlen=max_len, padding='pre')
  pred = model.predict(seq, verbose=0)
  pred_index = np.argmax(pred)
  return index_to_word[pred_index]
```

```python
seed_text = "what"
next_word = predictor(lstm_model, tokenizer, seed_text, max_len)
print(next_word)
```

```python
def generate_text(model, tokenizer, seed_text, max_len, num_words):
  for _ in range(num_words):
    next_word = predictor(model, tokenizer, seed_text, max_len)
    if next_word == "":
      break
    seed_text += " " + next_word
  return seed_text
```

```python
seed = "are a"
generate_text = generate_text(lstm_model, tokenizer, seed, max_len, 10)
print(generate_text)
```
